```python
import subprocess

result  = subprocess.run([
    "python", "seed-vc/inference.py",
    "--source", "ukranian_folk_music_voice_cloning/source/zhuk second voice.wav",
    "--target", "ukranian_folk_music_voice_cloning/target/ishov.wav",
    "--output", "ukranian_folk_music_voice_cloning/output",
    "--diffusion-steps", "50",
    "--length-adjust", "1.0",
    "--inference-cfg-rate", "0.7",
    "--f0-condition", "True",
    "--auto-f0-adjust", "False",
    "--semi-tone-shift", "0",
    "--config", "config_dit_mel_seed_uvit_whisper_base_f0_44k.yml",
    "--fp16", "True"
],
    capture_output=True,      # capture stdout and stderr
    text=True )
print("Return code:", result.returncode)
print("STDOUT:\n", result.stdout)
print("STDERR:\n", result.stderr)
```
